```python
natural = gp.read_file(gdrive+folder+"all_basins_no_dams/huc8_attributes.shp")
natural.columns
```

```python
b1920 = gp.read_file(gdrive+folder+"all_basins_before_1920/huc8_attributes.shp")
b1920
```

```python
natural.LENGTHKM_l
```

```python
b1920["frag_diff"] = b1920["LENGTHKM_l"] -  natural["LENGTHKM_l"]
```

```python
b1920.frag_diff.unique()
```

```python
b1920.to_file(gdrive+folder+"add_frag_diff/huc8_b1920_att.shp")
```

```python
b1950 = gp.read_file(gdrive+folder+"all_basins_before_1950/huc8_attributes.shp")
b1950["frag_diff"] = b1950["LENGTHKM_l"] -  b1920["LENGTHKM_l"]
b1950["frag_diff2"] = (b1950["LENGTHKM_l"] -  b1920["LENGTHKM_l"])+b1920["frag_diff"]
b1950.to_file(gdrive+folder+"add_frag_diff/huc8_b1950_att.shp")
```

```python
b1980 = gp.read_file(gdrive+folder+"all_basins_before_1980/huc8_attributes.shp")
b1980["frag_diff"] = b1980["LENGTHKM_l"] -  b1950["LENGTHKM_l"]
b1980["frag_diff2"] = (b1980["LENGTHKM_l"] -  b1950["LENGTHKM_l"])+b1920["frag_diff"]+b1950["frag_diff"]
b1980.to_file(gdrive+folder+"add_frag_diff/huc8_b1980_att.shp")
```

```python
unfil = gp.read_file(gdrive+folder+"all_basins_unfiltered/huc8_attributes.shp")
unfil["frag_diff"] = unfil["LENGTHKM_l"] -  b1980["LENGTHKM_l"]
unfil["frag_diff2"] = (unfil["LENGTHKM_l"] -  b1980["LENGTHKM_l"])+b1920["frag_diff"]+b1950["frag_diff"]+b1980["frag_diff"]
unfil.to_file(gdrive+folder+"add_frag_diff/huc8_unfil_att.shp")
```
